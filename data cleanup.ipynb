```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Read CSV 
df = pd.read_csv("..\Project3_Group1\Resources\Disasters.csv")
df
```

```python
# Drop irrelevant and keep relevant columns
keep = ['Year', 'Disaster Type', 'Disaster Subtype', 'Total Affected', "Total Damages ('000 US$)", "Country"]
df = pd.read_csv("..\Project3_Group1\Resources\Disasters.csv", usecols=keep)
df
```

```python
# Drop NaN values
df = df.dropna(how='any')
df
```

```python
# Drop data where year is before 1995
df1 = df[df['Year'] >= 1995]
df1
```

```python
# Renaming countries for readability and cohesiveness 
df1.loc[df['Country'] == 'United States of America (the)', 'Country'] = 'USA'
df1.loc[df['Country'] == 'Philippines (the)', 'Country'] = 'Philippines'
df1.loc[df['Country'] == 'Viet Nam', 'Country'] = 'Vietnam'
df1.loc[df['Country'] == 'Russian Federation (the)', 'Country'] = 'Russia'
df1.loc[df['Country'] == 'Turks and Caicos Islands (the)', 'Country'] = 'Turks and Caicos Islands'
df1.loc[df['Country'] == 'Iran (Islamic Republic of)', 'Country'] = 'Iran'
df1.loc[df['Country'] == 'Korea (the Republic of)', 'Country'] = 'North Korea'
df1.loc[df['Country'] == "Korea (the Democratic People's Republic of)", 'Country'] = 'South Korea'
df1.loc[df['Country'] == 'United Kingdom of Great Britain and Northern Ireland (the)', 'Country'] = 'UK'
df1.loc[df['Country'] == 'Bolivia (Plurinational State of)', 'Country'] = 'Bolivia'
df1.loc[df['Country'] == 'Bahamas (the)', 'Country'] = 'The Bahamas'
df1.loc[df['Country'] == 'Czech Republic (the)', 'Country'] = 'Czech Republic'
df1.loc[df['Country'] == 'Dominican Republic (the)', 'Country'] = 'Dominican Republic'
df1.loc[df['Country'] == 'Comoros (the)', 'Country'] = 'The Comoros'
df1.loc[df['Country'] == 'Taiwan (Province of China)', 'Country'] = 'Taiwan'
df1.loc[df['Country'] == 'Venezuela (Bolivarian Republic of)', 'Country'] = 'Venezuela'
df1.loc[df['Country'] == "Lao People's Democratic Republic (the)", 'Country'] = 'Laos'
df1.loc[df['Country'] == "Congo (the Democratic Republic of the)", 'Country'] = 'Democratic Republic of the Congo'
```

```python
# Renaming disaster subtype for readability 
df1.loc[df['Disaster Subtype'] == 'Land fire (Brush, Bush, Pasture)', 'Disaster Subtype'] = 'Land fire'
```

```python
# Output to csv 
file_path = ("output.csv")
df1.to_csv(file_path, index=False)
```
